# Which players under the age of 24 would be valuable strikers to purchase? 

**The Columns we will use to identify a valuable striker are:**

```python
# Creating a shortlist dataframe with only the columns we are considering in our criteria
shortlist_df = players_df[["sofifa_id", 
                        "long_name", 
                        "team_position", 
                         'club_name',
                        "age", 
                        "pace", 
                        "power_stamina", 
                        "shooting", 
                        "dribbling", 
                        "power_strength", 
                        'attacking_heading_accuracy',
                        'mentality_penalties',
                        "wage_eur", 
                        "value_eur", 
                        'contract_valid_until',
                        "release_clause_eur",
                        "overall", 
                        "potential",
                        'player_url'
                          ]]

# Printing the first 5 rows
shortlist_df.head(5)
```

```python
# Applying filters to 

shortlist_df = shortlist_df.loc[ # Filter to only focus on players below the age of 24
                                (shortlist_df["age"] <=24)
                                
                                # Filter to only focus on players who are pure strikers
                               & ((shortlist_df["team_position"] == "LS") | 
                                 (shortlist_df["team_position"] == "RS") |
                                 (shortlist_df["team_position"] == "ST") |
                                 (shortlist_df["team_position"] == "LF") |
                                 (shortlist_df["team_position"] == "CF") |
                                 (shortlist_df["team_position"] == "RF"))
                                
                                # Filter to only focus on players whose contract is expiring in or before 2023
                                & (shortlist_df["contract_valid_until"] <=2023.0)
    
                                # Filter to only focus on players who have a release clause below $40 Million
                                & (shortlist_df["release_clause_eur"] <= 40000000)
                               ]

print("Based on these filters we have", len(shortlist_df), "players to consider")
```

```python
shortlist_df.describe()
```

From the above dataframe we can observe what the median(50%) and 75% quartile values are. We will use these to further filter down on the players. 

- Pace must be 72
- Stamina must be 68
- Shooting must be 65 (Median) / 68 (75%)
- Dribbbling must be 65
- Power Strength should be 67

```python
shortlist_df = shortlist_df.loc[ (shortlist_df["pace"] >=72) 
                &  (shortlist_df["power_stamina"] >=68)
                &  (shortlist_df["shooting"] >=68)
                & (shortlist_df["dribbling"] >=65)
                & (shortlist_df["power_strength"]>=67)
                & (shortlist_df["overall"]>=70)].sort_values(["value_eur"])

print("Based on these filters we have", len(shortlist_df), "players to consider")
```

```python
# Adding a column to highlight if the player is a good header of the ball
shortlist_df.loc[(shortlist_df["attacking_heading_accuracy"] >=68), "good_header_of_the_ball"] = "Yes"
shortlist_df.loc[(shortlist_df["attacking_heading_accuracy"] <68), "good_header_of_the_ball"] = "No"
```

```python
# Printing out our shortlist of players
shortlist_df.sort_values("overall", ascending = False)
```

## Shubh's Formula
- ((0.5 * power_stamina) + Pace) * (Shooting + Dribbling)

```python
# Creating the S-Index Value for All Players
shortlist_df["s_index"] = (((0.5 * shortlist_df["power_stamina"]) + shortlist_df["pace"]) * (shortlist_df["shooting"] + shortlist_df["dribbling"]))/1000

# Rounding all values in our columns to show maximum of 1 decimal value
shortlist_df = shortlist_df.round(1)

# Keeping only the relevant columns for our final_shortlist
final_shortlist_df = shortlist_df[ ["sofifa_id", 
                            "long_name", 
                            "team_position",
                              "club_name",
                            "age", 
                            "wage_eur", 
                            "value_eur", 
                            'contract_valid_until',
                            "release_clause_eur",
                            "overall", 
                            "potential",
                            "s_index"]]


# Only keeping the top 20 "s_index" value players in our shortlist
final_shortlist_df = shortlist_df.sort_values("s_index", ascending = False).head(20)

# Printing our final short_list in descending order of s_index
final_shortlist_df
```

```python
# Median Score
shortlist_df.s_index.median() ### This means all the players in our final shortlist are above the median S_Index Score
```
